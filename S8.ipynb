### - Exercici 1

Agafa un conjunt de dades de tema esportiu que t'agradi i normalitza els atributs categòrics en dummy. Estandaritza els atributs numèrics amb StandardScaler.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler 
```

```python
wVb = pd.read_csv('cleanWVb.csv')
wVb.drop('Unnamed: 0', axis = 1, inplace = True)
wVb.sample(10)
```

We do not really need the names.

```python
wVb.drop('name', axis=1, inplace = True)
wVb.sample(5)
```

Let's make dummies of the positions

```python
dum = pd.get_dummies(wVb['position'])
dum.head()
```

Now, standarise numeric values.

```python
df = wVb.drop('position', axis = 1)

scale = StandardScaler()
stdData = scale.fit_transform(df)
df2 = pd.DataFrame(stdData, columns = df.columns)
#All means must be 0
print(df2.mean())
#All std must be 1
print(df2.std())
```

```python
finalDf = pd.concat([df2, dum], axis = 1)
finalDf
```

### Exercici 2

Continua amb el conjunt de dades de tema esportiu que t'agradi i aplica l'anàlisi de components principals.  
By applying PCA, we are building a new multidimensional space, where each individual will be placed. The position, or coordinates of this placement, in theory, better represent the differences between individuals, with the least amount of information.

```python
from sklearn.decomposition import PCA

n_components = 3
cols = [f'PCA_{i}' for i in range(n_components)]
pca = PCA(n_components = n_components)
pcaArray = pca.fit_transform(finalDf)
dfPCA = pd.DataFrame(pcaArray, columns = cols)
```

```python
exVar = pca.explained_variance_ratio_
cols = dfPCA.columns
for item in zip(cols, exVar):
    print(item[0], '=', f'{100*item[1]:.3f}%')

print('Explained total variance ratio', f'{exVar.sum()*100:.3f}%')
```

So in our case, we should maybe go higher than a 3 dimensional space.
